```python
import numpy as np
import pandas as pd

from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
```

```python
def load_txt(file_no):
    print('dataset ' + str(file_no))
    path_name = '/u/epw9kz/academic/first_year/ML/project/'
    
    if file_no == 1:
        file = path_name + 'dataset/project3_dataset1.txt'
    elif file_no == 2:
        file = path_name + 'dataset/project3_dataset2.txt'
    else:
        pass
        
    df = pd.read_csv(file, sep='\t', header=None)
    print(df.shape)

    for col in df.columns:
        if df[col].dtype == 'object':
            labelencoder = LabelEncoder()
            df[col] = labelencoder.fit_transform(df[col])

    df = df.values

    X = df[:, :-1]
    y = df[:, -1]
    
    return X, y
```

```python
X, y = load_txt(1) 
tsne = TSNE(n_components=2, random_state=0)

X_tsne = tsne.fit_transform(X)

# Plotting
plt.figure(figsize=(12, 8))

# Assuming binary classes for simplicity; adjust as needed
colors = ['blue', 'red']
labels = np.unique(y)
for color, label in zip(colors, labels):
    plt.scatter(X_tsne[y == label, 0], X_tsne[y == label, 1], c=color, label=label, alpha=0.5)

plt.title('t-SNE plot of dataset 1')
plt.xlabel('t-SNE dimension 1')  # X-axis title
plt.ylabel('t-SNE dimension 2')  # Y-axis title
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

X, y = load_txt(1) 
tsne = TSNE(n_components=3, random_state=0)

X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(20, 16))
ax = plt.axes(projection='3d')  # Create a 3D plotting axis

# Assuming binary classes for simplicity; adjust as needed
colors = ['blue', 'red']
labels = np.unique(y)

for color, label in zip(colors, labels):
    # We now also include the third dimension in the scatter plot
    ax.scatter(X_tsne[y == label, 0], X_tsne[y == label, 1], X_tsne[y == label, 2], c=color, label=label, alpha=0.5)

ax.set_title('3D t-SNE plot of dataset 1')
ax.set_xlabel('t-SNE dimension 1')
ax.set_ylabel('t-SNE dimension 2')
ax.set_zlabel('t-SNE dimension 3')  # Z-axis label for the third dimension
plt.legend()

plt.savefig('/u/epw9kz/academic/first_year/ML/project/dataset/images/tsne_3d_plot_data1.png', dpi=300, bbox_inches='tight')  # Saves the plot as a PNG file
plt.show()
```
